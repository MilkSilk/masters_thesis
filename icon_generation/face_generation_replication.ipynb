```python
faces = fetch_olivetti_faces()["data"]
```

```python
print(len(faces))
faces[0]
```

```python
for i in range(5):
    face_id = random.randint(0, 400)
    plt.figure()
    plt.imshow(faces[face_id].reshape(64,64), cmap="binary_r")
```

```python
model = LinearRegression()

n_independent = 150
random_noise = np.random.normal(0, 1, (len(faces), n_independent))
model.fit(random_noise, faces)
```

```python
for i in range(3):
    plt.figure()
    plt.imshow(model.predict(np.random.normal(0, 1, (1, n_independent))).reshape(64,64), cmap="binary_r")
```
